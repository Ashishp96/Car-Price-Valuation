#### 2.Remove Ask for price from price column and convert into int datatype.
#### 3.Remove comma and kms from kms_driven column and convert into int datatype.
#### 4.Slice upto 3 words in name column.

## Cleaning Data

### Year Column

```python
df=df[df['year'].str.isnumeric()]
```

```python
df['year']=df['year'].astype(int)
```

### Price Column

```python
df=df[df['Price']!='Ask For Price']
```

```python
df['Price']=df['Price'].str.replace(',','').astype(int)
```

### kms_driven has object values with kms at last 

```python
df['kms_driven']=df['kms_driven'].str.split().str.get(0).str.replace(',','')
```

```python
df=df[df['kms_driven'].str.isnumeric()]
```

```python
df['kms_driven']=df['kms_driven'].astype(int)
```

```python
df=df.dropna()
```

```python
df.info()
```

```python
df['name']=df['name'].str.split().str.slice(start=0,stop=3).str.join(' ')     
```

```python
df=df.reset_index(drop=True)
```

```python
df.to_csv('Cleaned car data.csv')
```

## Model Building

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe=OneHotEncoder()
ohe.fit(x[['name','company','fuel_type']])
```

```python
Column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),remainder='passthrough')
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(Column_trans,lr)
```

```python
pipe.fit(X_train,Y_train)
```

```python
Y_pred=pipe.predict(X_test)
```

```python
r2_score(Y_test,Y_pred)
```

```python
scores=[]
for i in range(1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(Column_trans,lr)
    pipe.fit(x_train,y_train)
    y_pred=pipe.predict(x_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```
